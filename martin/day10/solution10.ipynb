```python
def map_cell(x0,y0,s):
    x = 2*x0
    y = 2*y0
    if s == "|":
        c1 = (x,y-1)
        c2 = (x,y+1)
    elif s == "-":
        c1 = (x-1,y)
        c2 = (x+1,y)
    elif s == "L":
        c1 = (x,y-1)
        c2 = (x+1,y)
    elif s == "J":
        c1 = (x,y-1)
        c2 = (x-1,y)
    elif s == "7":
        c1 = (x,y+1)
        c2 = (x-1,y)
    elif s == "F":
        c1 = (x,y+1)
        c2 = (x+1,y)
    elif s == "S":
        return [((x,y),c) for c in ((x,y-1), (x,y+1), (x-1,y), (x+1,y))]
    elif s == ".":
        return []
    return [(c1,(x,y)), ((x,y),c2)]

```

```python
with open("input.txt") as fin:
    edges = set()
    start = None
    char_map = dict()
    for y, line in enumerate(fin):
        for x, s in enumerate(line.strip()):
            if s == "S":
                start = (2*x, 2*y)
            edges = edges.union(map_cell(x,y, s))
            char_map[(x,y)] = s
            
print("Start:", start)
print(edges)
```

```python
import networkx as nx

g = nx.Graph()
g.add_edges_from(edges)
cycle = nx.find_cycle(g, start)

print(len(cycle)//4)
```

```python
nodes_in_cycle = [e[0] for e in nx.find_cycle(g, start)]

min_x = min(e[0] for e in nodes_in_cycle)
max_x = max(e[0] for e in nodes_in_cycle)

min_y = min(e[1] for e in nodes_in_cycle)
max_y = max(e[1] for e in nodes_in_cycle)

print((min_x, max_x), (min_y, max_y))
```

```python
g2 = nx.Graph()
g2.add_edges_from(((x,y),n) for x in range(min_x,max_x+1) for y in range(min_y,max_y+1) if (x,y) not in nodes_in_cycle for n in ((x+1,y), (x, y+1)) if n not in nodes_in_cycle)
```

```python

cps = nx.connected_components(g2)
```

```python
import numpy as np
array = np.zeros(((max_y-min_y)//2+1, (max_x-min_x)//2+1))
for i, cp in enumerate(cps):
    if min_x < min([n[0] for n in cp]) and max([n[0] for n in cp]) < max_x and min_y < min([n[1] for n in cp]) and max([n[1] for n in cp]) < max_y:
        for (x,y) in cp:
            if min_x<=x<=max_x and min_y<=y<=max_y:
                if x % 2 == 0 and y % 2 == 0:
                    array[(y-min_y)//2,(x-min_x)//2] = 1
    
print(int(array.sum()))
```
